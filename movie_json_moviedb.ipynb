```python
https://api.themoviedb.org/3/movie/86990?api_key=51f02a795f9a3321ba1ed1deb79c9443
```

```python
tmdb_full=[]
```

```python

for ID1 in tmdb_ID_part:
    movies_url_imdb='https://api.themoviedb.org/3/movie/'+ID1+'?'+api_key
    try:
        r=requests.get(movies_url_imdb)
        if r.status_code !=200:
            continue
        else:
            full_data=r.json()
            tmdb_full.append(full_data)
    except:
        pass
df=pd.DataFrame(tmdb_full)
```

```python
full_TMDB_2020=pd.concat([part11,part12],axis=0)
```

```python
full_TMDB_2020.reset_index(drop=True,inplace=True)
```

```python
full_TMDB_2020.drop(columns='index',inplace=True)
```

```python
full_database=full_TMDB_2020.copy()
```

```python
full_database.to_csv('full_database.csv',index=False)
```

```python
full_database
```

```python
full_database.genres=full_database.genres.apply(lambda x:'|'.join(i['name'] for i in x))
```

```python
full_database.production_countries=full_database.production_countries.apply(lambda x:'|'.join(i['name'] for i in x))
```

```python
full_database.production_companies=full_database.production_companies.apply(lambda x:'|'.join(i['name'] for i in x))
```

```python
full_database.spoken_languages=full_database.spoken_languages.apply(lambda x:'|'.join(i['english_name'] for i in x))
```

```python
final=full_database.loc[:,['title','imdb_id','genres','release_date','spoken_languages','runtime','vote_average','vote_count','budget','revenue','production_countries','poster_path','id','original_language','original_title','overview','popularity','production_companies','tagline']]
```
